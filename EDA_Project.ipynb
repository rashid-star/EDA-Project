```python
# # Visualizing customer distribution by region
sns.countplot(x='region', data=df, color='r')
plt.title('Number of Customers per Region')
plt.show()
```

```python
# Defining the list of numeric columns to visualize
numeric_columns = ['age', 'bmi', 'children', 'charges']

# Plotting the distribution of each numeric column
# The KDE line shows the shape of the data (normal or skewed)
for col in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Visualizing the distribution of customers based on gender
sns.countplot(x=df['sex'])
plt.title('Gender Distribution of Customers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing the frequency of each unique value in the 'children' column
sns.countplot(x=df['children'])
plt.title('Count of Customers by Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing the number of smokers and non-smokers in the dataset
sns.countplot(x=df['smoker'], color='r')
plt.title('Smoking Status of Customers')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing pairwise relationships between numerical features
sns.pairplot(df, hue='smoker')
plt.show()
```

```python
# Creating boxplots for all numeric columns to check for outliers and data spread
for col in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
# Creating a heatmap to visualize correlation between numeric features
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
